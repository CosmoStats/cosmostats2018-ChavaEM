# Transformada de Fourier

La transformada de Fourier se define como:

$$\hat{f}(p) = {1 \over \sqrt{2 \pi}}\int_{-\infty}^{\infty}f(x) e^{-ipx} \, dx$$

Para hacer un codigo en Python, usamos la Transformada de Fourier Discreta (DFT):

$$X_k = \sum_{n = 0}^{N-1}x_n e^{-i2 \pi k n \, /\, N}$$

Que, esccrito en forma matricial, se puede expresar como:

$$\vec{X} = M \cdot \vec{x}$$

donde

$$M_{kn} = e^{-i2 \pi k n \, /\, N}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Funcion ara calcular la transformada de Fourier
def Fourier(x):
    x = np.asarray(x, dtype = float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    
    return np.dot(M, x)
```

```python
#Definimos la funcion gaussiana
def gauss(x, a, s, x0):
    f = x0 + a*np.exp(- (x/s)**2)
    return f
```

```python
x = np.arange(-60, 60, 0.1) 
a, s, x0 = 1, 1, 0   #Parametros para la funcion gaussiana
f = gauss(x, a, s, x0)

#Graficamos la funcion gaussiana
plt.plot(x, f)
plt.title("Funcion Gaussiana")
```

```python
#Graficamos la transformada de Fourier de la funcion gaussiana
plt.plot(x, Fourier(f))
plt.title("Transformada de Fourier (s = 1, x0 = 0)")
```

```python
x0 = 1 #Cambiamos x0 de 0 a 1
f = gauss(x, a, s, x0)
plt.plot(x, Fourier(f))
plt.title("Transformada de Fourier (s = 1, x0 = 1)")
```

```python
x0 = 0 
s = 0.1 #Bajamos la dispersion a 0.1
f = gauss(x, a, s, x0)
plt.plot(x, Fourier(f))
plt.title("Transformada de Fourier (s = 0.1, x0 = 0)")
```

```python
s = 10 #Aumentamos la dispersion a 10
f = gauss(x, a, s, x0)
plt.plot(x, Fourier(f))
plt.title("Transformada de Fourier (s = 0.1, x0 = 0)")
```

Al parecer, mientras mayor dispersión tiene la campana de Gauss, más angosta es su transformada de Fourier.

También pasa lo contrario, cuando la dispersión es muy pequeña en la Gaussiana, en la transformada de Fourier es mucho mayor.

```python
s = 1
f = gauss(x, a, s, x0)

#Graficamos la transformada de fourier, pero ahora con la libreria fft de numpy
plt.plot(x, np.fft.fft(f))
plt.title("Transformada de Fourier (s = 1, x0 = 0)")
```
