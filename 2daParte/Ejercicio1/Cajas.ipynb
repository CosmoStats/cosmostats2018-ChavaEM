```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
rcParams['figure.figsize'] = (8,8)
```

```python
lado = 1000 #Tama√±o de la caja
puntos = 2000 #Numero de puntos

#Coordenadas de los puntos aleatorios
x = [lado*(np.random.random() - 0.5) for i in range(puntos)]
y = [lado*(np.random.random() - 0.5) for i in range(puntos)]

#Grafica de los puntos en 2D
plt.plot(x, y, ',')
plt.gca().set_aspect('equal')
```

```python
lado3d = 1000 #Lado de la caja en 3D
puntos3d = 2000 #Numero de puntos en 3D

#Preparamos la grafica 3D
fig = plt.figure()
ax = Axes3D(fig)

#Generamos coordenadas para la grafica
x = [lado3d*(np.random.random() - 0.5) for i in range(puntos3d)]
y = [lado3d*(np.random.random() - 0.5) for i in range(puntos3d)]
z = [lado3d*(np.random.random() - 0.5) for i in range(puntos3d)]

#Graficamos
ax.scatter(x, y, z)
plt.gca().set_aspect('equal')
```
