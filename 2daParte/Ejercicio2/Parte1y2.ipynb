```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
import csv
from random import random

%matplotlib inline
rcParams['figure.figsize'] = (10,10)
```

### Caja con puntos en una malla cuadriculada

```python
lado = 1000 #Tamaño de la caja
xmalla, ymalla = [], []

#Puntos de la malla cuadriculada. Los guardamos en un archivo llamado malla.txt
for i in range(0, lado+1, 25):
    for j in range(0, lado+1, 25):
        xmalla.append(j)
        ymalla.append(i)

with open("malla.txt", "w") as malla:
    writer = csv.writer(malla, delimiter = '\t')
    writer.writerows(zip(xmalla,ymalla))

#Caja de malla cuadriculada
plt.plot(xmalla,ymalla, ',')
plt.gca().set_aspect('equal')
```

### Caja con puntos aleatorios en disco de radio 300

```python
lado = 1000 #Tamaño de la caja
radio = 300 #Radio del disco
puntos = 1000 #Numero de puntos en la caja

i = 0
xdisco, ydisco = [], []

#Generamos los puntos aleatorios en el disco
while i < puntos:
    
    #X y Y son las coordenadas de los puntos
    X = lado*np.random.random()
    Y = lado*np.random.random()
    
    if (X-lado*0.5)**2 + (Y-lado*0.5)**2 > radio**2: #Si las coordenadas estan fuera del circulo las descartamos
        continue
        
    #Agregamos al arreglo las coordenadas que esten dentro del disco
    xdisco.append(X)
    ydisco.append(Y)
    i += 1

with open("disco.txt", "w") as disco:
    writer = csv.writer(disco, delimiter = '\t')
    writer.writerows(zip(xdisco,ydisco))

#Graficamos el disco
plt.plot(xdisco,ydisco, '.')
plt.gca().set_aspect('equal')
plt.xlim(0, lado)
plt.ylim(0, lado)
```

```python
def anillo(R, r, c, puntos, lado): #Radio mayor, radio menor, centro, numero de puntos, lado de la caja
    x0, y0 = c
    i = 0 #Contador
    x, y = [], [] #Arreglos de las coordenadas
    
    #Generamos los puntos aleatorios en el anillo
    while i < puntos:
        
        #X y Y son las coordenadas de los puntos
        X = lado*np.random.random()
        Y = lado*np.random.random()
        
        if (X-x0)**2 + (Y-y0)**2 > R**2 or (X-x0)**2 + (Y-y0)**2 < r**2: #Si las coordenadas estan fuera del anillo las descartamos
            continue
            
        #Agregamos al arreglo las coordenadas que esten dentro del anillo y de la caja
        if (X < lado and X > 0) and (Y < lado and Y > 0):
            x.append(X)
            y.append(Y)
        i += 1
    #Graficamos el anillo
    return x, y
```

### Caja con puntos aleatorios en anillo de radios 290 y 300

```python
lado = 1000 #Tamaño de la caja
radioM = 300 #Radio mayor del anillo
radiom = 290 #Radio menor del anillo
puntos = 1000 #Numero de puntos en el anillo
c = (lado*0.5,lado*0.5) #Centro del anillo

xanillo, yanillo = anillo(radioM, radiom, c, puntos, lado)

with open("Anillo.txt", "w") as Anillo:
    writer = csv.writer(Anillo, delimiter = '\t')
    writer.writerows(zip(xanillo,yanillo))

plt.plot(xanillo,yanillo, 'b,')
plt.gca().set_aspect('equal')
plt.xlim(0, lado)
plt.ylim(0, lado)
```

```python
#Funcion que grafica una caja con puntos aeatorios

def caja_aleatoria(lado, Npuntos):
    #Coordenadas de los puntos
    X, Y = [lado * np.random.random() for i in range(Npuntos)], [lado * np.random.random() for i in range(Npuntos)]
    
    #Guardamos las coordenadas de los puntos por si los usamos despues
    with open("random.txt", "w") as random:
        writer = csv.writer(random, delimiter = '\t')
        writer.writerows(zip(X,Y))
    
    #Nos regresa los valores de las coordenadas de los puntos
    return X, Y
```

```python
lado_caja, Npuntos = 1000, 2000 #Tamaño de la caja y numero de puntos en ella

xrandom, yrandom = caja_aleatoria(lado_caja, Npuntos)

#Graficamos la caja con los puntos aleatorios
plt.plot(xrandom,yrandom, 'b.')
plt.gca().set_aspect('equal')
plt.xlim(0, lado_caja)
plt.ylim(0, lado_caja);
```

```python
#Funcion que calcula la distancia entre todos los puntos de una caja
def distancia(x1,y1,x2,y2):
    n = len(x1)
    d = [0 for i in range(int(n*(n-1)/2))]
    k = 0
    for i in range(n-1):
        for j in range(i+1, n):
            d[k] = (np.sqrt((x2[j]-x1[i])**2 + (y2[j]-y1[i])**2))
            k += 1
    return d
```

```python
d = distancia(xrandom, yrandom, xrandom, yrandom)
```

```python
Nbins = 35 #Numero de bins para el histograma
DD, bins_edges = np.histogram(d, bins=Nbins); #DD es el valor de los datos (aleatorios en este caso)

bins_edges = bins_edges[:-1] #Eliminamos la ultima posicion del bin para que DD y bins_edges sea de la misma longitud

#Graficamos el histograma de los puntos aleatorios
plt.hist(d, bins = Nbins);
plt.title("Histograma de distancias en caja de puntos aleatorios", fontsize = 16);
```

```python
#Definimos una funcion que calcule el estimador de Peebles-Hauser
def estimador_PH(DD, Nbins, lado, Npuntos):
    Xrandoms, Yrandoms = caja_aleatoria(lado, Npuntos) #Tomamos una muestra aleatoria
    dist = distancia(Xrandoms, Yrandoms, Xrandoms, Yrandoms) #Obtenemos las distancias de esa muestra aleatoria
    
    RR, bin_puntos = np.histogram(dist, bins = Nbins) #Obtenemos los datos que se usaran para calcular el estimador
    
    return DD/RR - 1 #Calculamos el estimador
```

```python
estim = estimador_PH(DD, Nbins, lado_caja, Npuntos) #Obtenemos el estimador para DD datos aleatorios
```

```python
#Graficamos el estimador
plt.plot(bins_edges, estim)
plt.hlines(0, 0, np.sqrt(2)*lado_caja)
plt.ylim(-1,1)
plt.title("Estimador de Peebles-Hauser para DD puntos aleatorios", fontsize = 16);
```

Podemos ver que para distancias pequeñas el estimador es cercano a cero. Para distancias más grandes, éste diverge, pues los bins son muy pequeños y el cociente entre ellos suele variar mucho.
