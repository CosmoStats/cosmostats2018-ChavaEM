```python
from random import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
lenght = 10000
x = np.array([random() for i in range(lenght)])
y = np.array([random() for i in range(lenght)])
X, Y =[], []
r = 0.5
for i in range(len(x)):
    if (x[i]-0.5)**2+(y[i]-0.5)**2 <= r**2:
        X.append(x[i])
        Y.append(y[i])
plt.plot(X,Y, ',')
plt.gca().set_aspect('equal')
```

```python
R = x*np.cos(y)+0.5
theta = x*np.sin(y)+0.5
plt.plot(R,theta, ',')
plt.gca().set_aspect('equal')
```
