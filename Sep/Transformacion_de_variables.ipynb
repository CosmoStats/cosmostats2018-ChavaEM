```python
from random import random
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
lenght = 4000
x = np.array([2*random() - 1 for i in range(lenght)])
y = np.array([2*random() - 1 for i in range(lenght)])
X, Y =np.array([]), np.array([])
r = 1
for i in range(len(x)):
    if x[i]**2+y[i]**2 <= r**2:
        X = np.append(X,x[i])
        Y = np.append(Y,y[i])
plt.plot(X,Y, ',')
plt.gca().set_aspect('equal')
```

```python
r = np.absolute(x)
theta = pi * y
plt.plot(r * np.cos(theta), r * np.sin(theta), ',')
plt.gca().set_aspect('equal')
```

```python
r2 = np.sqrt(r)
plt.plot(r2 * np.cos(theta), r2 * np.sin(theta), ',')
plt.gca().set_aspect('equal')
```

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.array([2*random() - 1 for i in range(lenght)])
y = np.array([2*random() - 1 for i in range(lenght)])
z = np.array([2*random() - 1 for i in range(lenght)])
X, Y, Z =np.array([]), np.array([]), np.array([])
r = 1
for i in range(len(x)):
    if x[i]**2 + y[i]**2 +z[i]**2 <= r**2:
        X = np.append(X,x[i])
        Y = np.append(Y,y[i])
        Z = np.append(Z, z[i])
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X,Y,Z, marker= '.')
plt.gca().set_aspect('equal')
```

```python
R = np.absolute(x)
Theta = pi/2 * y
Phi = pi * z
SinT = np.sin(Theta)
CosT = np.cos(Theta)
SinP = np.sin(Phi)
CosP = np.cos(Phi)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(R*SinT*CosP, R*SinT*SinP, R*CosT, marker= '.')
plt.gca().set_aspect('equal')
```
