# Distribución uniforme. Transformación de variables.
En este notebook vamos a generar círculos y esferas a partir de distribuciones uniformes de dos maneras distintas.

```python
#Importamos librerias que vamos a ocupar
from random import random
from pylab import rcParams
import numpy as np
import matplotlib.pyplot as plt
from math import pi

%matplotlib inline
rcParams['figure.figsize'] = (10, 6)
```

```python
#Vamos a utilizar 4000 datos.
lenght = 4000

#Preparamos los arreglos asignando valores aleatorios en [-1,1] en coordenadas cartesianas
x = np.array([2*random() - 1 for i in range(lenght)])
y = np.array([2*random() - 1 for i in range(lenght)])
r = 1
#Agregamos los datos que estan dentro del circulo unitario en un nuevo arreglo
for i in range(len(x)):
    if x[i]**2+y[i]**2 <= r**2:
        X = np.append(X,x[i])
        Y = np.append(Y,y[i])

#Graficamos ese nuevo arreglo
plt.plot(X,Y, ',')
plt.gca().set_aspect('equal')
```

```python
#Generamos valores aleatorios en coordenadas polares. r en [0,1] y Theta en [-pi/2, pi/2]
r = np.absolute(x)
theta = pi * y
plt.plot(r * np.cos(theta), r * np.sin(theta), ',')
plt.gca().set_aspect('equal')
```

```python
from mpl_toolkits.mplot3d import Axes3D

#Generamos valores aleatorios en coordenadas cartesianas en 3D, todas entre [-1,1]
x = np.array([2*random() - 1 for i in range(lenght)])
y = np.array([2*random() - 1 for i in range(lenght)])
z = np.array([2*random() - 1 for i in range(lenght)])
X, Y, Z =np.array([]), np.array([]), np.array([])
r = 1

#Incluimos solamente los valores que estan dentro de la esfera unitaria dentro de otro arreglo y lo graficamos
for i in range(len(x)):
    if x[i]**2 + y[i]**2 +z[i]**2 <= r**2:
        X = np.append(X,x[i])
        Y = np.append(Y,y[i])
        Z = np.append(Z, z[i])
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X,Y,Z, marker= '.')
plt.gca().set_aspect('equal')
```

```python
#Generamos los valores aleatorios en coordenadas esfericas. R en [0,1], Theta en [-pi/2,pi/2] y Phi en [-pi,pi]
R = np.absolute(x)
Theta = pi/2 * y
Phi = pi * z
SinT = np.sin(Theta)
CosT = np.cos(Theta)
SinP = np.sin(Phi)
CosP = np.cos(Phi)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(R*SinT**2*CosP, R*SinT**2*SinP, R*CosT*SinT, marker= '.')
plt.gca().set_aspect('equal')
```
