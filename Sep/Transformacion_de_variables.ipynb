```python
from random import random
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
lenght = 10000
x = np.array([2*random() - 1 for i in range(lenght)])
y = np.array([2*random() - 1 for i in range(lenght)])
X, Y =np.array([]), np.array([])
r = 1
for i in range(len(x)):
    if x[i]**2+y[i]**2 <= r**2:
        X = np.append(X,x[i])
        Y = np.append(Y,y[i])
plt.plot(X,Y, ',')
plt.gca().set_aspect('equal')
```

```python
r = np.absolute(x)
theta = pi * y
plt.plot(r * np.cos(theta), r * np.sin(theta), ',')
plt.gca().set_aspect('equal')
```

```python
r2 = np.sqrt(r)
plt.plot(r2 * np.cos(theta), r2 * np.sin(theta), ',')
plt.gca().set_aspect('equal')
```
