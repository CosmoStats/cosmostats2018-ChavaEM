# Función de distribución acumulada para la distribución Gaussiana

A partir de la distribución Gaussiana:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}}exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

podemos encontrar la funciń de distribución acumulada definida como:

$$F(x) = \int_{- \infty}^x f(x')dx'$$

Al sustituir la distribución gaussiana y haciendo el cambio de variable: $u = \frac{x-\mu}{\sqrt{2}\sigma}$ con $du = \frac{dx}{\sqrt{2}\sigma}$, tenemos:

$$F(x) = \frac{1}{\sqrt{\pi}} \int_{- \infty}^{\frac{x-\mu}{\sqrt{2}\sigma}}e^{-u^2}du = \frac{1}{\sqrt{\pi}}\left[ \int_{- \infty}^0 e^{-u^2}du + \int_0^{\frac{x-\mu}{\sqrt{2}\sigma}}e^{-u^2}\right]du$$

Dado que:

$$\int_{-\infty}^0 e^{-u^2}du = \frac{\sqrt{\pi}}{2}$$

y

$$erf(x) = \frac{2}{\sqrt{\pi}}\int_0^x e^{-u^2}du$$

Tenemos que:

$$F(x) = \frac{1}{2}\left[ 1 + erf\left(\frac{x-\mu}{\sqrt{2}\sigma}\right) \right]$$

## Gráficas de distribución Gaussiana y CDF

Importamos librerías necesarias

```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
from math import sqrt, pi

%matplotlib inline
```

Graficamos la distribución Gaussiana

```python
#Declaramos los parametros mu y sigma de nuestra distribucion
mu = 70
sigma = 2

#Hacemos los arreglos de la variable independiente, x, y la distribucion, f.
x = np.arange(60,80, 0.1)
f = 1/(sqrt(pi * 2) * sigma) * np.exp(-(x - mu)**2/(2 * sigma**2))
plt.plot(x,f)
plt.grid(True)
```

Graficamos la función de distribución acumulada $F(x)$, de acuerdo al resultado obtenido arriba (justo antes del código)

```python
F = 1/2 * (1 + sp.erf((x - mu)/(sqrt(2) * sigma)))
plt.plot(x,F)
plt.grid(True)
```

Ahora vamos a generar números aleatorios entre 0 y 1 y haciendo uso de la función de distribución acumulada (CDF), obtener el valor correspondiente de x. Para ello vamos a ocupar la función inversa de la CDF, despejando simplemente x de la ecuación para $F(x)$:

$$\frac{x - \mu}{\sqrt{2} \sigma} = erf^{-1}(2F - 1)$$

obteniendo:

$$x = \sqrt{2} \sigma \, erf^{-1}(2F-1) + \mu$$

Para tener la misma notación que las notas, cambiaremos $x$ por $H_0$:

$$H_0 = \sqrt{2} \sigma \, \, erf^{-1}(2y-1) + \mu$$

```python
from random import random

#generamos los numeros aleatorios y los asignamos a y en un arreglo numpy
y = np.array([random() for i in range(1000)])

#Calculamos el H0 correspondiente con la ecuacion de arriba (H0 es un numpy array)
H0 = sqrt(2) * sigma * sp.erfinv(2 * y - 1) + mu
#Graficamos el histograma
n, bins, patches = plt.hist(H0, 50, facecolor = 'g')
plt.xlabel('H_0')
plt.ylabel('Frecuencia')
plt.title('Histograma de H_0')
plt.grid(True)
```

Podemos notar que a simple vista, parece que el histograma sigue una distribución Gaussiana, tal como se pensaba, pues esperábamos obtener la distribución con la que iniciamos.
